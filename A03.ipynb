# Part A

```python
import numpy as np 
import pandas as pd 
import seaborn as sns

# read the data
c = pd.read_csv("cigarette.csv")
c
```

```python
# get average by states
avg = c.groupby(by = ["state"], as_index = False).mean()

# get top 5 states and store as a list
top_5 = avg.sort_values(by = ['price'], ascending = False).head(5)
top_5_list = top_5['state'].tolist()
```

```python
# filter data using the top 5 states list
a = c[c["state"].isin(top_5_list)]
a
```

```python
# print the output
_ = a.boxplot(by ='state', column =['price'])
```

# Part B

```python
import datetime
import matplotlib.pyplot as plt
%matplotlib inline

# read the data
df = pd.read_csv('Pandemic.csv')
df
```

```python
# understand the data: look up unique values
df['location'].unique()
```

```python
# work on European time format
df['ds'] = pd.to_datetime(df['ds'], dayfirst=True)
```

```python
# Country 1: South Korea
df_sk = df[df['location'].isin(['South Korea'])]

table_sk = pd.pivot_table(df_sk, values='value', index=['ds', 'location'],
                    columns=['data_type'], aggfunc=np.sum).reset_index(level='location')

_ = table_sk.plot(kind='line', title = "Condition of Covid-19 cases in South Korea", xlabel = "Months", ylabel = "Covid cases")
```

```python
# Country 2: Ethiopia
df_et = df[df['location'].isin(['Ethiopia'])]

table_et = pd.pivot_table(df_et, values='value', index=['ds', 'location'],
                    columns=['data_type'], aggfunc=np.sum).reset_index(level='location')

_ = table_et.plot(kind='line', title = "Condition of Covid-19 cases in Ethiopia", xlabel = "Months", ylabel = "Covid cases")
```

```python
# Task 1
n = pd.read_csv("Nobel Prize.csv")
n.head()
```

```python
# top ten countries by count of prizes
nc = n.groupby(["Birth Country"])["Prize"].count().reset_index(name="Count")

nc.sort_values(by = ["Count"], ascending = False).head(10)
```

```python
# Task 2

# check female value from sex column
n['Sex'].unique()
```

```python
# create female dataframe

nf = n.loc[(n['Sex'] == 'Female')]
nf
```

```python
# find the full name when the year is the smallest
nf['Full Name'].loc[(nf['Year'] == nf['Year'].min())]
```

```python
# Task 3
# check the variables
n['Category'].unique()
```

```python
# print category, sex, count
ns = n.groupby(['Category','Sex'])["Prize"].count().reset_index(name="Count")
ns
```

```python
# render the output where male is much dominant in every single categories
sns.catplot(x='Category', y='Count', hue='Sex', kind = 'bar', data=ns)
```

```python
# Task 4
# print top 25 organizations by count
no = n.groupby(["Organization Name"])["Prize"].count().reset_index(name="Count")

no.sort_values(by = ["Count"], ascending = False).head(25)
```

```python
# Task 5
# Explore possible data: Comparison of individual and organizational winners by years 
nr = pd.DataFrame(n, columns = ['Year', 'Laureate Type'])
nr
```

```python
# render stripplot
sns.stripplot(x="Laureate Type", y="Year", data=nr)
```

```python
# Part D
# read the data and filter required information
covid = pd.read_csv("df.csv")
covid = covid.groupby(['Country_Region'])[['Confirmed','Recovered']].sum().reset_index()
covid = covid.sort_values(by = ["Confirmed"], ascending = False).head(5)
covid.head()
```

```python
# render the output

# Figure Size
fig, ax = plt.subplots(figsize = (16, 9))
 
# Horizontal Bar Plot
ax.barh(covid['Country_Region'], covid['Confirmed'], color ='r', label='Confirmed')
ax.barh(covid['Country_Region'], covid['Recovered'], color ='g', label='Recovered')
 
# Show top values
ax.invert_yaxis()
 
# Add annotation to bars
for i in ax.patches:
    plt.text(i.get_width()+0.2, i.get_y()+0.3,
             i.get_width(),
             size = 12,
             color ='k')

plt.legend()
plt.xlabel("Recovered")
plt.ylabel("Country")


# Show Plot
plt.show()
```
