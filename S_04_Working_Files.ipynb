```python
# Example 2: Lets extract ages from the given string
ages = re.findall(r'\d{1,3}', nameage)
#findall() matches all occurrences of a pattern
# r'\d{1,3}: This will return two digit numbers,
        #which means, looking for two digit number starting from {1,3} (starting position 1 and end at 3 (excluding 3))
ages
```

```python
# Example 3: Lets extract the names only from the string
names = re.findall(r'[A-Z][a-z]*',nameage)
#r'[A-Z][a-z]*': Return string containing capital letter [A-Z] and small letters [a-z] and 
               #'*' - will match 1 or more uppercase or lowercase letters
# will need to match patterns as well

names
```

```python
# Example 4: Lets create a dictionary for names and respective ages
agedict = {}
x = 0
for eachname in names:
    agedict[eachname]=ages[x]
    x+=1 #doing two tasks at one time
print(agedict)
#  Adds a value and the variable and assigns the result to that variable
```

```python
# Step 1: Read the "script01.txt" and "script02.txt"
file1 = open("script01.txt", "r")
file2 = open("script02.txt", "r")

script1 = file1.read()
script2 = file2.read()

script =script1 + script2
script
```

```python
# Step 3: Create your own function 
for alphabet in script:
    al = alphabet.lower()
    if al in atoz:
        if al not in outcome:
            outcome[al] = 1
        else:
            outcome[al] += 1
    
outcome_sort = sorted(outcome.items(), reverse = True, key = lambda x: x[1])
```

### What is Train/Test

+ Train/Test is a method to measure the accuracy of your model.

   + It is called Train/Test because you split the the data set into two sets: a training set and a testing set.
   + 80% for training, and 20% for testing.
   + You train the model using the training set.
   + You test the model using the testing set.
   + Train the model means create the model.
   + Test the model means test the accuracy of the model.
   + As we work with datasets, a machine learning algorithm works in two stages.
+ We usually split the data around 20%-80% between testing and training stages.
+ Under supervised learning, we split a dataset into a training data and test data in Python ML.


+ Separating data into training and testing sets is an important part of evaluating data mining models. ... By using similar data for training and testing, you can minimize the effects of data discrepancies and better understand the characteristics of the model.

## Detecting Multicollinearity Using Variance Inflation Factors (VIF)

+ Multicollinearity occurs when there are two or more independent variables in a multiple regression model, which have a high correlation among themselves.
+ When some features are highly correlated, we might have difficulty in distinguishing between their individual effects on the dependent variable.
+ Multicollinearity can be detected using various techniques, one such technique being the **Variance Inflation Factor(VIF).**


```python
# Task 10: Detecting Multicollinearity Using Variance Inflation Factors
from statsmodels.stats.outliers_influence import variance_inflation_factor
def get_vif_factors( X ):
    X_matrix = X.to_numpy()
    vif = [ variance_inflation_factor( X_matrix, i ) for i in range( X_matrix.shape[1] ) ]
    vif_factors = pd.DataFrame()
    vif_factors['column'] = X.columns
    vif_factors['vif'] = vif
    return vif_factors
```

```python
vif_factors = get_vif_factors( X[X_features])
vif_factors
```

```python
columns_with_large_vif = vif_factors[vif_factors.vif > 4].column
columns_with_large_vif
```
