```python
import random

# generate a random number between 1 and 100
rand_num = random.randint(1,100)
print("I got a number between 1 and 100.")

# user starts with 100 points
user_p = 100
print("You start with", user_p, "points. For each guess, 1 point will be deducted.")

# compare the numbers using Try Except statement
while True:
    
  try:
    # user input
    guess_num = int(input("Guess the number: "))

    if guess_num <1 or guess_num>100:
      print("Provide a number from 1 to 100.")
    elif guess_num > rand_num:
      user_p = user_p - 1      
      print("Too high")
    elif guess_num < rand_num:
      user_p = user_p - 1
      print("Too low")
    elif guess_num == rand_num:
      user_p = user_p - 1
      print("Bingo! Your final score is ", user_p)
      break


  except Exception:
    print("Your input is not valid.")
```

```python
grade_list=[]

count=int(input("How many students do you need to enter grades for: "))
order=0

while count>0:
  order=order+1
  name=input("Enter the name of student {}: ".format(order))
  grade=input("Enter the student's final grade: ")
  subset=[name,grade]
  grade_list.append(subset)
  count=count-1

print(grade_list)

textfile = open("grades.txt", "w")
for subset in grade_list:
    for i in subset:
        textfile.write(i + "\n")
textfile.close()
```
